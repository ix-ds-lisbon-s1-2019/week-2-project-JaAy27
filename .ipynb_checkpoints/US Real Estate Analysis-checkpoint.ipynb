## Data Visualization of US Real Estate

### Data Description

The dataset was found at https://www.kaggle.com/goldenoakresearch/us-acs-mortgage-equity-loans-rent-statistics. It covers variables such as the monthly mortgage, debt, home owner costs of certain locations in the States. The research was conducted by Golden Oak and published in 2017. It gives an overall analysis of the U.S. Real Estate sector.
Please note that some columns were removed for simplicity and efficiency reasons.

### Data Dictionary  

A full dictionary can be found in the file ***real_estate_document.pdf*** but here is a general overview of the main concepts:

Note: They are all floats except for the variable state, which is a string.

- UID: The ID of the location analyzed.
- state (string): The name of the State.
- latitude: Latitude of the area.
- longitude: Longitude of the area.
- pop: Male and female population of the area.
- second_mortgage: Percent of houses with a second mortgage.
- debt: Percentage of homes with some type of debt.
- pct_own: Percentage of people owning a property.
- hi_mean: The mean household income of the specified geographic location.
- rent_mean: Mean of the contract rent plus the estimated average monthly cost of utility features
- rent_stdev: The standard deviation of the gross rent for the specified geographic location.
- hs_degree: Percentage of people with at least a high school degree.
- hc_mortgage_mean: The mean Monthly Mortgage and Owner Costs of specified geographic location.
- hc_mean: The mean Monthly Owner Costs of specified geographic location.
- home_equity: Percentage of homes with a home equity loan.
- married: The percentage of the population who are married.
- separated: The percentage of the population who are separated.
- divorced: The percentage of the population who are divorced.

### Data Loading

```python
from datascience import *
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy import stats
import seaborn as sns
from ipywidgets import interact
import warnings
warnings.filterwarnings("ignore")
```

```python
re = pd.read_csv('real_estate_db.csv')
```

```python
re.head()
```

### Overview

```python
import pandas_profiling

report = pandas_profiling.ProfileReport(re)
```

```python
report
```

The report shows us that some variables should be ignored because of their high correlation with other variables. Thus, we choose to analyze the variables that are not too correlated with other variables and that would be interesting for the analysis.

```python
#Removing the rejected columns for simplicity
re = re.drop(['family_mean', 'family_median', 'family_stdev', 'female_pop', 'male_pop', 'hc_median', 'rent_median'], axis=1)
```

We will fill the null values with 0.

```python
re = re.fillna(0)
```

### Duplicates

```python
re[re.duplicated()].shape
```

```python
#Removing the duplicated rows
re = re.drop_duplicates()
re.shape
```

### Variable Grouping

First, we want to determine which states have the highest rent. To do so, we will start by grouping the values by state.

```python
re_state = re.groupby(['state'], as_index=False).mean()
```

We will save it into a new csv, which we will use in later steps.

```python
re_state.to_csv('re_state.csv', index=False)
```

## Visualizations

### Rent by State

The main goal of the visualizations is to paint a picture of the Real Estate sector of the US. We will first make a bar plot to have an idea of the distribution by state.

```python
%matplotlib inline
re_sort = re_state.sort_values('rent_mean')
```

```python
re_sort.plot.barh('state', 'rent_mean', figsize=(9,9));
```

If we look at the bar plot, we can see that Hawaii is noticeably more expensive than the rest. California and District of Columbia follow up. If we map the rent mean, we can see the geographical distribution of the rent.

```python
re_map = Table.read_table('re_state.csv')
```

```python
rent = re_map['rent_mean']
colors = []

for item in rent:
        if item > 1000:
            color = 'darkred'
            colors.append(color) 
        elif item <800 :
            color = 'beige'
            colors.append(color)
        else:
            color = 'lightred'
            colors.append(color)
            

rent_map = re_map.with_columns('Color', colors)
new_map = rent_map.select('latitude', 'longitude', 'state', 'Color')
```

```python
Marker.map_table(new_map)
```

##### The darkest color represents the highest average rent and the lightest one, the lowest. We can see that the coastal states tend to have higher rent.

### Variable Correlation

```python
def interactive_plot(col1, col2):
    sns.lmplot(x=col1, y=col2, data=re_state);
```

```python
interact(interactive_plot, col1=re_state.columns.drop('state'), col2=re_state.columns.drop('state'));
```

From the above interactive plot, a few observations regarding the US Real Estate stand out:

- The regions with more populations tend to have younger inhabitants (pop vs male_age_mean or female_age_mean)
- the rent mean and debt have a rather strong positive correlation.
- The rent mean and hc_mortgage_mean have a strong positive correlation.
- Debt and percentage of ownership (pct_own) have a slightly negative correlation.
- The average household income and the divorce rate have a strong negative correlation. (hi_mean vs divorced)
- The average household income and debt have a rather strong positive correlation.
- The average household income and percentage ownership (pct_own) have a slightly negative correlation.

## Conclusions

- From the visualizations, we can conclude that the coastal states tend to have higher rent. 


- The areas with higher rent mean also tend to have higher mortgage and ownership costs and more indebted people. However, those with greater proportions of indebted people tend to have lower percentage ownership, which could mean that their debt is not necessarily related to mortgage.   


- In terms of demographics, the more populated regions tend to have younger inhabitants and the areas with higher average household income tend to have higher debt but lower divorce rate.


- Finally, an interesting observation is that the areas with higher percentage of house ownership tend to have lower household income mean. The richest ones might prefer renting a property in coastal areas.
